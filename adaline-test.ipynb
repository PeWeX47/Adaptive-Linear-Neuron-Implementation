```python
from sklearn.model_selection import train_test_split
from adaline import Adaline
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.metrics import accuracy_score
```

```python
sns.set()
```

```python
def generate_linear_data(num_samples=100):
    X = 2 * np.random.rand(num_samples, 2) - 1
    y = 2 * (X[:, 0] - 0.5 * X[:, 1] + 0.2) + 0.1 * np.random.randn(num_samples)
    y = np.where(y >= 0, 1, -1)
    return X, y
```

```python
X, y = generate_linear_data(num_samples=1000)
```

```python
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], c='red', label='Class -1')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='blue', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('2D data')
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
adaline = Adaline(eta=0.0001, epochs=500, is_verbose=True)            
adaline.fit(X_train, y_train)
plt.scatter(range(adaline.epochs), adaline.list_of_errors)
plt.xlabel('Epochs')
plt.ylabel('MSE')
```

```python
y_pred = adaline.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
def generate_3d_linear_data(num_samples=100):
    X = 2 * np.random.rand(num_samples, 3) - 1
    y = 2 * (X[:, 0] - 0.5 * X[:, 1] + 0.2 * X[:, 2]) + 0.1 * np.random.randn(num_samples)
    y = np.where(y >= 0, 1, -1)
    return X, y
```

```python
X, y = generate_3d_linear_data(num_samples=1000)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2', 'Feature 3'])
df['Class'] = np.where(y == -1, 'Class -1', 'Class 1')

fig = px.scatter_3d(df, x='Feature 1', y='Feature 2', z='Feature 3', color='Class')

fig.update_layout(title='3D data')
fig.update_traces(marker_size = 4)

fig.show()
plt.show()
```

```python
adaline = Adaline(eta=0.0001, epochs=500, is_verbose=True)            
adaline.fit(X_train, y_train)
plt.scatter(range(adaline.epochs), adaline.list_of_errors)
plt.xlabel('Epochs')
plt.ylabel('MSE')
```

```python
y_pred = adaline.predict(X_test)
accuracy_score(y_test, y_pred)
```
