```python
from sklearn.model_selection import train_test_split
from adaline import Adaline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score
```

```python
def generate_linear_data(num_samples=100):
    np.random.seed(42)
    X = 2 * np.random.rand(num_samples, 2) - 1
    y = 2 * (X[:, 0] - 0.5 * X[:, 1] + 0.2) + 0.1 * np.random.randn(num_samples)
    y = np.where(y >= 0, 1, -1)
    return X, y
```

```python
X, y = generate_linear_data(num_samples=1000)
```

```python
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], c='red', label='Class -1')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='blue', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear data')
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
adaline = Adaline(eta=0.0001, epochs=500, is_verbose=True)            
adaline.fit(X_train, y_train)
plt.scatter(range(adaline.epochs), adaline.list_of_errors)
plt.xlabel('Epochs')
plt.ylabel('MSE')
```

```python
y_pred = adaline.predict(X_test)
accuracy_score(y_test, y_pred)
```
